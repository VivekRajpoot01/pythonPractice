### 1. Relational Plot
- to see the statisticaal relation between 2 or more variables
- Bivariate Analysis
#####   Plots under this section
     - scatterplot
     - lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px  ## to load only tha datesets
```

```python
# loading the tips dataset
tips=sns.load_dataset('tips')
tips
```

```python
## scatter plot  -> axes label function
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
# used hue parameter to define each person in graph
```

```python
# relplot -> figure label function  -> square shape
#import seaborn as sns
#import matplotlib.pyplot as plt

#sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
sns.relplot(data=tips, x='total_bill', y='tip', kind= 'scatter',hue='sex',style='time',size='size')

```

```python
# line plot
# we can use the dataset gapminder from plotly about life expectancy
gap=px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country']=='India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df, x='year',y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
## getting three countries data
temp_df1=gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df1
```

```python
sns.relplot(kind='line',data=temp_df1, x='year',y='lifeExp',hue='country')
```

```python
## when we use figure level function then the legend will be outside
## the graph and vice-versa for axis level
```

```python
## facet plot -> only work with figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex', row='smoker')
```

### 2. Distributiion Plot
- used for univariate analysis
- used to find out the distribution
- range of the observation
- Centreal Tendency
- Is the data bimodel?
- Are there Outlier
###### plots Under Distribution Plot
- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins parameter
sns.histplot(data=tips,x='total_bill',bins=20)
```

```python
# It is also possible to visualize the distribuion of a categorical variable using the logic of histogram.
# Discrete bins are automatically set for categorical data
```

```python
# actually acting as a countplot 
sns.histplot(data=tips,x='day')
```

```python
# hue parameter
sns.displot(data=tips, x='tip',kind='hist',hue='sex')
```

```python
# faceting using col and row -> not work with histplot function
sns.displot(data=tips, x='tip',kind='hist',col='sex',element='step')
```

#### Kde Plot
- Kernal Density estimator
- rather than using discrete bins, a kde plot smooths the observations with a Gaussian Kernel, producing a continuous density estimate

```python
tips
```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
# Rugplot

# plot marginal distribution by drawing ticks along the x and y axes

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
#  Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observation within each rectangle with the fill order

sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
# Bivariate kdeplot
# a bivariate kde plot smooths the (x,y) observation with a 2D Gaussian
sns.kdeplot(data=tips,x='total_bill',y='tip')
```
