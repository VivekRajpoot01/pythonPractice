```python
batsman = pd.read_csv('sharma-kohli.csv')
```

```python
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# Using Multiple plot
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title so that easilt understand what graph want to show
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma and Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#FFC733') # or we can provide hex code for color value
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title("Rohit Sharma and Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# line style and width
plt.plot(batsman['index'],batsman['V Kohli'],color='#FFC733',linestyle='solid',linewidth='5') # linestyle may be dotted dashed dashdot 
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth='3')
plt.title("Rohit Sharma and Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker and marker width
plt.plot(batsman['index'],batsman['V Kohli'],color='#FFC733',linestyle='solid',linewidth='5',marker='o',markersize='10') # to mark points in the graph 
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth='3',marker='D')
plt.title("Rohit Sharma and Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# to label individual line in the graph to show which line belong to which person
plt.plot(batsman['index'],batsman['V Kohli'],color='#FFC733',linestyle='solid',linewidth='5',marker='o',markersize='10',label='Virat Kohli') #  
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth='3',marker='D',label='Rohit Sharma')
plt.title("Rohit Sharma and Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
# to show the labelling in the graph we call the function plt.legend()
plt.legend()
plt.grid()
plt.show()   # to show the graph in some editor graph directly not shown then we have to use plt.show()
```

```python
## we can also trim the graph 
## by using plt.xlim() and plt.ylim()
```

```python
# logarithmic scale
# load numpy array
arr = np.load('big-array.npy')
plt.hist(arr)
```

```python
## From above we can see that in the range of 30-40 there
## too much please present in it hence it ignoring some values present in 10-20 scale
## TO AVOID THIS WE CAN USE LOGARITHMIC SCALE
plt.hist(arr,bins=[10,20,30,40,50,60,70],log=True)
##  now we can see some data between 10-20 is present
```

## Pie Chart
1. Univariate / Bivariate Analysis 
2. Categorical vs Numerical
3. Use Case - To find contribution on a standard scale

```python
# simple data
data = [23,45,60,89,75]
subjects = ['Maths','English','Science','SST','Hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
## using dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
## creating pie chart
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
## to show percentage by each batsman
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
## we can explode one section or more from the pie chart by using explode
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0.3,0.0001
                                                                           ,0],shadow=True)
plt.show()
```
