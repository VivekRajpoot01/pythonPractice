### Statistical Practical Implementation
#### Measure of Central Tendency
- Mean
- Median
- Mode

```python
## let assume
ages=[10,45,20,26,35,14,22,31,100]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
## there is no impact of outlier on the median
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))
```

```python
import seaborn as sns
sns.boxplot(ages)
```

#### Number Summary

```python
import numpy as np
q1,q3=np.percentile(ages,[25,75])
```

```python
## to check outlier [lower fence - Higher fence]
iqr = q3-q1
lower_fence = q1-(1.5*iqr)
higher_fence = q3+(1.5*iqr)
print(lower_fence,higher_fence)
```

```python
print(q1,q3)
```

#### Measure of Dispersion
1. Variance
2. Standard Deviation

```python
statistics.variance(ages)  ## use sample varinace formulat (n-1)
```

```python
np.var(ages)  ## use population variance formule (n)
```

```python
## calculate varinace by making own function
def variance(data):
    n=len(ages)
    ## mean of the data
    mean=sum(data)/n
    ## variance
    deviation=[(x-mean)**2 for x in data]
    variance = sum(deviation)/n
    return variance
```

```python
variance(ages)
```

```python
## population varinace using statistical
statistics.pvariance(ages)
```

```python
## finding standard deviation
import math
sd = math.sqrt(statistics.pvariance(ages))
sd
```

#### Check whether distribution is Normal Distribution

```python
## If we want to check whether feature is gaussian or normal distributed
## Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```
